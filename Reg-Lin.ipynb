```python
import pandas as pd
import numpy as np
import csv
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Location = r'C:\Users\ThinkYoga\Desktop\MLOT17\reglin\regLin.csv'
df = pd.read_csv(Location)
```

```python
X_train , X_test, Y_train, Y_test = train_test_split(df[df.columns[0:1]],df[df.columns[1:2]],test_size=0.25, random_state=42) 
```

```python
plt.scatter(X_train,Y_train)
```

```python
linReg=LinearRegression()
linReg.fit(X_train,Y_train)
```

# Train model

```python
print linReg.intercept_,linReg.coef_
```

```python
np.mean((linReg.predict(X_test) - Y_test)**2)
```

# Visualize

```python
x = np.linspace(X_test['X'].min(),X_test['X'].max(),1000)
y = x*linReg.coef_[0]+linReg.intercept_[0]
```

```python
plt.scatter(X_test,Y_test)

plt.plot(x,y,color='red')
```

# Graficamos el error de Wo

```python
wo = linReg.intercept_[0]
rango = np.linspace(wo-20,wo+20,100)
errores = []
for wo in rango:
        predicted = X_test['X']*linReg.coef_[0] + wo
        errores += [np.mean( (predicted - Y_test['y'])**2)]

plt.plot(rango,errores)
```
